## 389. Find the Difference

You are given two strings `s` and `t`.

String `t` is generated by random shuffling string `s` and then add one more letter at a random position.

Return the letter that was added to `t`.

 

**Example 1:**

Input: s = "abcd", t = "abcde"  
Output: "e"  
Explanation: 'e' is the letter that was added.  

**Example 2:**

Input: s = "", t = "y"  
Output: "y"  

```python
def findTheDifference(s, t):
    table = {}
    for letter in s:
        if letter in table:
            table[letter] += 1
        else:
            table[letter] = 1
    for letter in t:
        if letter not in table:
            return letter
        elif letter in table:
            table[letter] -= 1

        if table[letter] == -1:
            return letter
```

```python
s = "abcd"
t = "abcde"
findTheDifference(s,t)
```

```python
s = ""
t = "y"
findTheDifference(s, t)
```

```python
s = "a"
t = "aa"
findTheDifference(s, t)
```

```python
def findTheDifference2(s, t):
    list_t = list(t)
    for letter in s:
        list_t.remove(letter)
    return list_t[0]
```

```python
s = "abcd"
t = "abcde"
findTheDifference2(s,t)
```

```python
s = ""
t = "y"
findTheDifference2(s, t)
```

```python
s = "a"
t = "aa"
findTheDifference2(s, t)
```
